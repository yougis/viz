## Cleanning du dataset

```python
data["date_adhesion"] = pd.to_datetime(data["Date d'adhésion"],format='%d/%m/%Y')
```

```python
data.loc[data['nb Adhérents'] > 1, "montant_adhesion"] = '4000'
data.loc[data['nb Adhérents'] > 1, "famille"] = 1
data.loc[data['nb Adhérents'] == 1, "montant_adhesion"] = '2000'
data.loc[data['nb Adhérents'] == 1, "indiv"] = 1

cols = ["famille", "indiv","nb Adhérents"]
data[cols] = data[cols].fillna(0).astype(int)
data.head()
```

### Ajout d'une métode renvoyant le dataframe filtré à partir d'une date.

```python
class StatDav(param.Parameterized):
    
    # WIDGETS ------------------------------------------------
    # slider date
    date_slider = param.Integer(data["date_adhesion"].min().year,  bounds=(data["date_adhesion"].min().year, data["date_adhesion"].max().year))
    

    # METHODES ------------------------------------------------
    # création de la metode qui renvoit le dataframe filtré
    @param.depends('date_slider')
    def select_date(self):
        return data.loc[data["date_adhesion"].dt.year == self.date_slider]
    
    @param.depends('date_slider')
    def getDfFromYear(self):
        filter = data["date_adhesion"].dt.year >= self.date_slider
        df_year = data[filter]
        return df_year[["NOM","Prénom","date_adhesion","famille","indiv","nb Adhérents","montant_adhesion"]]

```

### Ajout d'un widget et d'une métode renvoyant le dataframe filtré sur le type d'adhésion

### Ajout d'un widget et d'une métode renvoyant le nombre d'adhésion correspondant aux différents choix des widgets précédents

```python
def formatter(value):
    return str(value)

class StatDav(param.Parameterized):    
  
   # WIDGETS ------------------------------------------------
    # slider date
    date_slider = param.Integer(data["date_adhesion"].min().year,  bounds=(data["date_adhesion"].min().year, data["date_adhesion"].max().year))
    
    date_range_slider = param.Range((data["date_adhesion"].min().year,data["date_adhesion"].max().year),  bounds=(data["date_adhesion"].min().year, data["date_adhesion"].max().year))
      
    # Radio bouton    
    type_adhesion = param.Selector(objects=['famille', 'indiv'])
    
    montant_adhesion_individuel = param.Integer(2000, bounds=(100,3000) )
    montant_adhesion_famille = param.Integer(4000, bounds=(100,5000))
    
    ratio_famille_indiv = param.Number( default=0.3,  bounds=(0.1,1))

    static_ad_indiv = 2000
    static_ad_famille = 4000
    
        
    # METHODES ------------------------------------------------
    # création de la metode qui renvoit le dataframe filtré
    @param.depends('date_slider','date_range_slider')
    def select_date(self):        
        return data.loc[data["date_adhesion"].dt.year == self.date_slider]
    
    @param.depends('date_slider','date_range_slider')
    def getDfFromYear(self):
        rangeTuple = self.date_range_slider
        filter = (data["date_adhesion"].dt.year >= rangeTuple[0])&(data["date_adhesion"].dt.year <= rangeTuple[1])
        df_year = data[filter]
        return df_year[["NOM","Prénom","date_adhesion","famille","indiv","nb Adhérents","montant_adhesion"]]

    
    @param.depends('date_slider','date_range_slider')
    def yearStat(self):
        df = self.getDfFromYear()
        statDict = {
            'adhesion_indiv' : len(df.loc[(df["indiv"]==1)]), 
            'adhesion_famille' : len(df.loc[(df["famille"]==1)]),
            'total' : len(df.loc[(df["indiv"]==1)]) +  len(df.loc[(df["famille"]==1)]),
            'adhérent' : int(df.loc[(df["famille"]==1)]["nb Adhérents"].sum()) + len(df.loc[(df["indiv"]==1)]),
            'ratio' :  round(len(df.loc[(df["famille"]==1)]) / (len(df.loc[(df["indiv"]==1)]) +  len(df.loc[(df["famille"]==1)])),2)
        }
        return statDict.get('total'), statDict.get('adhesion_indiv'), statDict.get('adhesion_famille'),  statDict.get('adhérent'), statDict.get('ratio')
   
    @param.depends('date_slider','date_range_slider')
    def butifulValues(self):
        tuple = self.yearStat()
        tresor = self.getTresor()
        return pn.pane.Markdown("""
        # De/A l'année %s / %s """ %self.date_range_slider + """
        ## %s adhésions
        
        - %s adhésions individuelles 
        - %s adhésion famille 
        
        ## %s adhérents 
        
        il y a %s adhésions famille pour 1 adhésion individuelle 
        
        """ %tuple +
        """## %s Francs
        """ %tresor 
                               )
    
    @param.depends('date_slider','date_range_slider','type_adhesion','date_range_slider')
    def getDfByTypeAdhesion(self):
        df = self.getDfFromYear()
        newDf= df.loc[(df[self.type_adhesion]==1)]
        return newDf
    
    @param.depends('date_slider','type_adhesion','date_range_slider')
    def getStat(self):
        df = self.getDfByTypeAdhesion()
        return len(df)
    
    @param.depends('date_slider','montant_adhesion_individuel','montant_adhesion_famille','date_range_slider')
    def getTresor(self):
        statDf = self.yearStat()
        montant_tresor = statDf[1] * self.montant_adhesion_individuel + statDf[2] * self.montant_adhesion_famille
        return montant_tresor
    
     
    @param.depends('date_slider','date_range_slider')
    def bar(self):
        df =  self.getDfFromYear()
        famille = df.loc[(df["famille"]==1)]
        bar = famille.hvplot.hist(y='nb Adhérents').opts(title="Nombre d'adhérent par famille",ylabel="Nb adhésion famille", yticks=range(1,50),xticks=[1,2,3,4,5,6],padding=1)
        moyenne = famille["nb Adhérents"].mean()
        moyenne_line = hv.VLine(moyenne)
        layout = bar * moyenne_line
        layout.opts(opts.VLine(color='red', line_width=4))
        return layout 
    
    @param.depends('date_slider','montant_adhesion_individuel','montant_adhesion_famille', 'ratio_famille_indiv','date_range_slider')
    def line(self):
        dfm = self.calculateDataset()
        statTuple =  pd.DataFrame(list(self.yearStat()))
        lineCalc = dfm.hvplot.line(x="nb_adhesion", y="tresor_calc_ratio_famille", yformatter=formatter).opts(title=" Modelisation de l'impact du prix d'adhésion", ylabel="Tresor", xlabel="Nombre d'adhésion", xlim=(0,150), ylim=(0,350000))
        
        lineYear =  hv.Points((statTuple.loc[1]+statTuple.loc[2] , (statTuple.loc[1]* self.static_ad_indiv ) + (statTuple.loc[2]* self.static_ad_famille) )).opts(size=8)

        layout = lineCalc *lineYear
        return layout
   
    @param.depends('montant_adhesion_individuel','montant_adhesion_famille', 'ratio_famille_indiv')
    def tresorCalc(self):
        df = self.calculateDataset()
        tresor = int(df["tresor_calc_ratio_famille"].loc[100])
        return tresor

    @param.depends('montant_adhesion_individuel','montant_adhesion_famille', 'ratio_famille_indiv')
    def tresorCalcMarkdow(self):
        tresor = self.tresorCalc()
        return pn.pane.Markdown(""" # %s Francs pour 100 adhésions """ % tresor, width=1000)
        
    @param.depends('montant_adhesion_famille', 'ratio_famille_indiv','montant_adhesion_individuel')
    def calculateDataset(self):
        dfm = pd.DataFrame({'nb_adhesion': range(1,300)})
        df =  self.getDfFromYear()
        
        famille = df.loc[(df["famille"]==1)]
        moyenne = famille["nb Adhérents"].mean()
        dfm["nb_adherents"] =  dfm['nb_adhesion'] * moyenne
        dfm["nb_adherents_calc"] = (dfm["nb_adherents"] * self.ratio_famille_indiv) + (dfm['nb_adhesion'] *(1- self.ratio_famille_indiv))
        dfm["tresor_calc_ratio_famille"] = round(((dfm['nb_adhesion'] * self.ratio_famille_indiv )* self.montant_adhesion_famille)+ (dfm['nb_adhesion'] *(1- self.ratio_famille_indiv)* self.montant_adhesion_individuel ))
        return dfm

    @param.depends('date_slider','montant_adhesion_individuel','montant_adhesion_famille', 'ratio_famille_indiv','date_range_slider')
    def calculEcartTresor(self):
        statDf = self.yearStat()
        df = self.calculateDataset()
        tresorCalc = int(df["tresor_calc_ratio_famille"].loc[statDf[0]])
        montant_tresor = statDf[1] * self.static_ad_indiv + statDf[2] * self.static_ad_famille
        ecart = tresorCalc -  montant_tresor
        return pn.pane.Markdown(
            """ 
             Montant du trésors réel : **%s** Francs 
            """ % montant_tresor +
            """ 
             Montant du trésors calculé : **%s** Francs
            """ % tresorCalc +
            """ ### **%s** Francs d'écart """ % ecart, width=1000)
   
    @param.depends('date_slider','ratio_famille_indiv','date_range_slider')
    def calculRatioReel(self):
        statDf = self.yearStat()
        total = statDf[0]
        famille = statDf[0] * self.ratio_famille_indiv
        indiv = total - famille
        return pn.pane.Markdown(
           """
           %s adhésions familles
           """ % int(famille) +
           """
           %s adhésions individuelles
           """ % int(indiv)
        )
    
    @param.depends('date_slider','ratio_famille_indiv','date_range_slider')
    def adhesion_adherent(self):
        statDf = self.yearStat()
        df = self.calculateDataset()
        adherent = int(df["nb_adherents_calc"].loc[statDf[0]])
        line = df.hvplot.line(x='nb_adhesion',y='nb_adherents_calc').opts(xlim=(0,150), ylim=(0,300))
        return line
    
    @param.depends('date_slider','ratio_famille_indiv','date_range_slider')
    def adhesion_adherentMarkdown(self):
        statDf = self.yearStat()
        df = self.calculateDataset()
        adherent = int(df["nb_adherents_calc"].loc[statDf[0]])
        return pn.pane.Markdown(
           """
           %s adhérents
           """ % int(adherent))
```

```python
stat = StatDav(name="Statistique des adhésions de DAV")


```

```python
app =  pn.Column(pn.Column(pn.Param(stat.param['date_range_slider']),pn.Row(stat.butifulValues, stat.bar)))

app2 =  pn.Row(
    pn.Row(
        pn.Column(
           # pn.Param(stat.param['date_range_slider']),
            pn.Param(stat.param['montant_adhesion_individuel']),
            pn.Param(stat.param['montant_adhesion_famille']),
            pn.Param(stat.param['ratio_famille_indiv'])
        ),
            
        pn.Column(
            pn.Row(stat.line),
            stat.adhesion_adherent
        ),                  
            
        pn.Column(
            pn.Spacer(margin=10),
            stat.tresorCalcMarkdow, 
            stat.calculEcartTresor, 
            stat.calculRatioReel,
            stat.adhesion_adherentMarkdown
        )
    )
)
app.append(app2)
app.show()
```
